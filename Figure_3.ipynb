```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Step 1: Load the data
data = {
    'Months': [
        "Phase_1", "Phase_2", "Phase_3", "Phase_4", "Phase_5", 
        "Phase_6", "Phase_7", "Phase_8", "Phase_9", "Phase_10", 
        "Phase_11", "Phase_12"
    ],
    'Method_A_Cost': [3000, 3000, 3000, 3000, 3000, 3000, 3000, 3000, 3000, 3000, 3000, 3000],
    'Method_B_Cost': [126, 126, 126, 126, 126, 126, 126, 126, 126, 126, 126, 126],
}

df = pd.DataFrame(data)

# Step 2: Calculate Total Costs for Each Method
df['Total_Cost_Method_A'] = df['Method_A_Cost']
df['Total_Cost_Method_B'] = df['Method_B_Cost']

# Step 3: Visualize Results
plt.plot(df['Months'], df['Total_Cost_Method_A'], label='Method A', color='skyblue')
plt.plot(df['Months'], df['Total_Cost_Method_B'], label='Method B', color='orange')

# Annotate the last value for Method A
plt.annotate('£{:,.0f}'.format(df['Total_Cost_Method_A'].iloc[-1]), 
             xy=(len(df['Months']) - 1, df['Total_Cost_Method_A'].iloc[-1]), 
             xytext=(len(df['Months']), df['Total_Cost_Method_A'].iloc[-1] - 100),
             color='blue')

# Annotate the last value for Method B
plt.annotate('£{:,.0f}'.format(df['Total_Cost_Method_B'].iloc[-1]), 
             xy=(len(df['Months']) - 1, df['Total_Cost_Method_B'].iloc[-1]), 
             xytext=(len(df['Months']), df['Total_Cost_Method_B'].iloc[-1] - 50),
             color='red')

plt.xlabel('Reanalysis Phases')
plt.ylabel('Cost (GBP)')
plt.legend(loc='upper left')

# Adding vertical lines and aligning labels with them
x_positions = np.arange(len(df['Months']))
for x_pos in x_positions:
    plt.axvline(x=x_pos, color='black', linestyle='--', linewidth=0.5)

# Set crossed x-axis labels
plt.xticks(ticks=x_positions, labels=df['Months'], rotation=45, ha='right')

# Set y-axis ticks: multiples of 100 up to 500, then 1000
y_ticks = np.concatenate((np.arange(0, 501, 250), np.arange(1000, 4001, 1000)))
plt.yticks(y_ticks)

plt.grid(axis='x', linestyle='--', linewidth=0.5, color='black', alpha=0.7)

plt.tight_layout()
plt.show()

# Step 4: Output results
print("Total Cost for Method A:", df['Total_Cost_Method_A'].sum())
print("Total Cost for Method B:", df['Total_Cost_Method_B'].sum())
```
