Static costs Method A, B | Costs/trio

```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Load the data
data = {
    'Months': [18, 36, 54, 72, 90, 108, 126, 144, 162, 180, 198, 216],
    'Method_A_Cost': [3000, 3000, 3000, 3000, 3000, 3000, 3000, 3000, 3000, 3000, 3000, 3000],
    'Method_B_Cost': [500, 500, 500, 500, 500, 500, 500, 500, 500, 500, 500, 500],
}

df = pd.DataFrame(data)

# Step 2: Calculate Total Costs for Each Method
df['Total_Cost_Method_A'] = df['Method_A_Cost']
df['Total_Cost_Method_B'] = df['Method_B_Cost']

# Step 3: Visualize Results
plt.plot(df['Months'], df['Total_Cost_Method_A'], label='Method A', color='blue')
plt.plot(df['Months'], df['Total_Cost_Method_B'], label='Method B', color='red')
plt.xlabel('Months')
plt.ylabel('Cost (GBP)')
plt.legend()

# Adding vertical lines and labels for months
months = [18, 36, 54, 72, 90, 108, 126, 144, 162, 180, 198, 216]
for month in months:
    plt.axvline(x=month, color='black', linestyle='--', linewidth=0.5)

plt.xticks(months)
plt.show()

# Step 4: Output results
print("Total Cost for Method A:", df['Total_Cost_Method_A'].sum())
print("Total Cost for Method B:", df['Total_Cost_Method_B'].sum())
```
